```python
%matplotlib inline
import matplotlib
import cv2
from matplotlib.pyplot import imshow 

import numpy as np
from imgaug import augmenters as iaa
```

```python
# load example image 
image = np.load('example.npy')
```

# Original "image"

```python
imshow(image.reshape(480,480)[140:360,140:360], cmap='Greys')
```

# Rotated

```python
a = iaa.Affine(rotate=(180))

imshow(a.augment_images(image.reshape(1,480,480)).reshape(480,480)[140:360,140:360], cmap='Greys',)
print("\t Paper")
```

# Dropout


```python
a = iaa.Dropout(p=0.42)

imshow(a.augment_images(image.reshape(1,480,480)).reshape(480,480)[140:360,140:360], cmap='Greys',)
```

# Gaussian blur

```python
a = iaa.GaussianBlur(sigma=3.2)

imshow(a.augment_images(image.reshape(1,480,480)).reshape(480,480)[140:360,140:360], cmap='Greys',)
```

# Gaussian noise

```python
a = iaa.AdditiveGaussianNoise(loc=0.03, scale=255*0.3)

imshow(a.augment_images(image.reshape(1,480,480)).reshape(480,480)[140:360,140:360], cmap='Greys',)
```

# Elastic

```python
a = iaa.ElasticTransformation(alpha=2.1)

imshow(a.augment_images(image.reshape(1,480,480)).reshape(480,480)[140:360,140:360], cmap='Greys',)
```
